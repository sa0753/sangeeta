```
import urllib.request
from bs4 import BeautifulSoup

weather_url="https://forecast.weather.gov/MapClick.php?textField1=33.22&textField2=-97.15#.Xp_jD8hKjIV"
open_page = urllib.request.urlopen(weather_url)
weather_soup=BeautifulSoup(open_page)
#print (weather_soup)


seven_day_forecast = weather_soup.find(id="seven-day-forecast")


period_weather = seven_day_forecast.select(".tombstone-container .period-name")
period = [i.get_text() for i in period_weather]
short_descsription = [b.get_text() for b in seven_day_forecast.select(".tombstone-container .short-desc")]
temperature = [f.get_text() for f in seven_day_forecast.select(".tombstone-container .temp")]
descsription = [x["title"] for x in seven_day_forecast.select(".tombstone-container img")]

import pandas as pd
weather_forecast = pd.DataFrame({
    "period": period,
    "short_descsription": short_descsription,
    "temp": temperature,
    "descsription":descsription
})
print(weather_forecast)

weather_forecast.to_csv('data.csv')
```
